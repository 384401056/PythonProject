```python
''' 
Statsmodels 是用来探索数据、估计模型，并运行统计检验的Python包。
'''
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family='SimHei', size=14)
%matplotlib inline
import seaborn as sns
from numpy._distributor_init import NUMPY_MKL

df = pd.read_csv(r'C:\Users\Administrator\Desktop\iris.data', names=['sepal length', 'sepal width', 'petal length', 'petal width','class'])
```

```python
#  在这里，让我们使用它来构建一个简单的线性回归模型.为steosa类中花萼长度和花萼宽度之间的关系进行建模。

fig, ax = plt.subplots(figsize=(7, 7))
ax.scatter(df['sepal width'][:50], df['sepal length'][:50])
ax.set_ylabel('Sepal 长度')
ax.set_xlabel('Sepal 宽度')
ax.set_title('Setosa Sepal Width vs. Sepal Length', y = 1.02)
```
