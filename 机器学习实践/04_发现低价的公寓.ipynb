```python
# 去除数据中的单位,并转换数据类型
df['面积'] = df['面积'].map(lambda x: float(str(x).split('平')[0]))
df['年代'] = df['年代'].map(lambda x: int(str(x).split('年')[0]))
df['房屋总价'] = df['房屋总价'].map(lambda x: float(str(x).split('万')[0]))
df['参考首付'] = df['参考首付'].map(lambda x: float(str(x).split('万')[0]))
df['房屋单价'] = df['房屋单价'].map(lambda x: float(str(x).split('元')[0]))
df['参考月供'] = df['参考月供'].map(lambda x: float(str(x).split('元')[0]))
# min(df['年代'])
# df.groupby('房屋单价')['面积'].unique().to_frame()

# # 装修程度为'毛坯'的集合
# mu = df[df['装修程度'].str.contains('毛坯')]
# su = df[df['装修程度'].str.contains('精装修')]
# su
# df.columns
# df.head().T
```

```python
# 筛选出房型为'4室'的数据框.
mu = df[df['房型'].str.contains('4室')]
# 筛选出房型为非'2卫'的数据框.
mu = df[~(df['房型'].str.contains('2卫'))]
mu
```

```python
def parse_address(row):
    if '－' in row:
        area, temp, street = row.split('－')[:3]
        comm = temp.strip().lstrip().rstrip('\t') #　去除数据中的\t\t\t\t\t
        street = street.strip().lstrip().rstrip('\t')
    else:
        area = ''
        comm = ''
        street = ''
    return pd.Series({'行政区':area, '社区':comm, '街道':street})

attr = df['位置'].apply(parse_address)
attr

# df['行政区'] = df['位置'].apply(parse_address)
# df['社区'] = df['位置'].apply(parse_address)
# df['街道'] = df['位置'].apply(parse_address)
```

```python
# 将attr的字段加入df数据框中
su = df.join(attr) 
su

# mu = su[su['社区'].str.contains('翠湖')]
# mu
# min(mu['房屋单价'])
```
